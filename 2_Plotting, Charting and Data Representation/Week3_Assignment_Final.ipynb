```python
# Animation

n = 100

# generate 4 random variables from the random, gamma, exponential, and uniform distributions
x1 = np.random.normal(0, 1, n)
x2 = np.random.gamma(2, 1.5, n)
x3 = np.random.exponential(2, n)
x4 = np.random.uniform(0,5, n)

def update(curr):
    # check if animation is at the last frame, and if so, stop the animation a
    if curr == n: 
        a.event_source.stop()
    
    b = 10
    
    #ax1.set_ylim(0, 0.4)
    ax1.hist(x1[:curr], bins=b, color="tomato")
    ax1.set_title("Normal, n = {}".format(curr))
    ax2.hist(x2[:curr], bins=b, color="aquamarine")
    ax2.set_title("Gamma")
    #ax2.set_xlim(0,15)
    ax3.hist(x3[:curr], bins=b, color="lawngreen")
    ax3.set_title("Exponential")
    #ax3.set_xlim(7,20)
    ax4.hist(x4[:curr], bins=b, color="gold")
    ax4.set_title("Uniform")
    #ax4.set_xlim(14,20)
    
    #ax1.legend(loc = 4, frameon = false, title = 'n = {}'.format(curr))
    
    fig.tight_layout(h_pad=1, w_pad = 1)
    
    return ax1, ax2, ax3, ax4

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharey = True)

a = animation.FuncAnimation(fig, update, interval=100)
```
