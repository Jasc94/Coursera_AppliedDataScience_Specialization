# Pandas Idioms

```python
import pandas as pd
import numpy as np
import timeit

# with '.' I tell Python that the file is in the folder I am
df = pd.read_csv('./Course1_Resources/resources/week-3/datasets/census.csv')
df.head()
```

```python
# Chaining

# Pull out state and city as multiple index
# With the () we tell Python that we will expand the code in several lines
(df.where(df['SUMLEV'] == 50)
    .dropna()
    .set_index(['STNAME', 'CTYNAME'])
    .rename(columns = {'ESTIMATESBASE2010' : 'Estimates Base 2010'}))
```

```python
# More traditional, non-pandorable way
df = df[df['SUMLEV'] == 50]
df.set_index(['STNAME', 'CTYNAME'], inplace = True)
df.rename(columns = {'ESTIMATESBASE2010' : 'Estimates Base 2010'})
```

```python
# Which method runs faster?
```

```python
def first_approach():
    global df
    
    return (df.where(df['SUMLEV'] == 50)
            .dropna()
            .set_index(['STNAME', 'CTYNAME'])
            .rename(columns = {'ESTIMATESBASE2010' : 'Estimates Base 2010'}))

# Read in our dataset anew
df = pd.read_csv('./Course1_Resources/resources/week-3/datasets/census.csv')

# let's run it and test it
timeit.timeit(first_approach, number = 10)
```

```python
def second_approach():
    global df
    
    new_df = df[df['SUMLEV'] == 50]
    new_df.set_index(['STNAME', 'CTYNAME'], inplace = True)
    return new_df.rename(columns = {'ESTIMATESBASE2010' : 'Estimates Base 2010'})

df = pd.read_csv('./Course1_Resources/resources/week-3/datasets/census.csv')

timeit.timeit(second_approach, number = 10)
```

```python
# The second approach is much faster

# Stylistic idioms might have performance issues
```

```python
# Map function
# Pandas has something similar called 'applymap'
# Pandas also has a function called 'apply'
```

```python
# We'll create a function that finds the min and max within the data

def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    
    return pd.Series({'min' : np.min(data), 'max' : np.max(data)})
```

```python
# We call capply in the dataframe

df.apply(min_max, axis = 'columns').head(20)
```

```python
# Revised version adding two new columns in the original dataframe to store min and max

def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    
    # New entry for max
    row['max'] = np.max(data)
    
    # New entry for min
    row['min'] = np.min(data)
    
    return row

df.apply(min_max, axis = 'columns')
```

```python
# Using lambdas

rows = ['POPESTIMATE2010','POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2013',
                'POPESTIMATE2014', 'POPESTIMATE2015']

df.apply(lambda x: np.max(x[rows]), axis = 1).head()
```

```python
# Divide the state into four categories: Northeast, Midwest, Midwest, South

def get_state_region(x):
    northeast = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 
                 'Rhode Island','Vermont','New York','New Jersey','Pennsylvania']
    midwest = ['Illinois','Indiana','Michigan','Ohio','Wisconsin','Iowa',
               'Kansas','Minnesota','Missouri','Nebraska','North Dakota',
               'South Dakota']
    south = ['Delaware','Florida','Georgia','Maryland','North Carolina',
             'South Carolina','Virginia','District of Columbia','West Virginia',
             'Alabama','Kentucky','Mississippi','Tennessee','Arkansas',
             'Louisiana','Oklahoma','Texas']
    west = ['Arizona','Colorado','Idaho','Montana','Nevada','New Mexico','Utah',
            'Wyoming','Alaska','California','Hawaii','Oregon','Washington']
    
    if x in northeast:
        return 'Northeast'
    elif x in midwest:
        return 'Midwest'
    elif x in south:
        return 'South'
    else:
        return 'West'
    
```

```python
# We set the apply function

df['state_region'] = df['STNAME'].apply(lambda x: get_state_region(x))
```
