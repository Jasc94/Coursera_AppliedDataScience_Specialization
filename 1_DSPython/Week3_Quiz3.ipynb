## Q1

```python
# First, we create two DataFrames: staff and students
staff_df = pd.DataFrame([{'Name' : 'Kelly', 'Role' : 'Director of HR'},
                         {'Name' : 'Sally', 'Role' : 'Course liasion'},
                         {'Name' : 'James', 'Role' : 'Grader'}])

# Let's index these by staff name
staff_df = staff_df.set_index('Name')

# Now we create the student DataFrame
student_df = pd.DataFrame([{'Name' : 'James', 'School' : 'Business'},
                           {'Name' : 'Mike', 'School' : 'Law'},
                           {'Name' : 'Sally', 'School' : 'Engineering'}])

# Lets index this one by name too
student_df = student_df.set_index('Name')
```

```python
df = pd.merge(student_df, staff_df, how = 'outer', left_index = True, right_index = True)
df
```

```python
df = pd.merge(student_df, staff_df, how = 'left', left_index = True, right_index = True)
df
```

```python
df = pd.merge(student_df, staff_df, how = 'right', left_index = True, right_index = True)
df
```

### Answer:
df = pd.merge(student_df, staff_df, how = 'left', left_index = True, right_index = True)

```python
import pandas as pd

df = pd.DataFrame(['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D'],
                  index = ['excellent', 'excellent', 'excellent', 'good', 'good', 'good', 'ok', 'ok', 'ok', 'poor', 'poor'], columns = ['Grades'])

my_categories = pd.CategoricalDtype(categories = ['D', 'D+', 'C-', 'C', 'C+', 'B-', 'B', 'B+', 'A-', 'A', 'A+'],
                                  ordered = True)

grades = df['Grades'].astype(my_categories)

result = grades[(grades>'B') & (grades<'A')]
result
```

### Answer:
my_categories = pd.CategoricalDtype(categories = ['D', 'D+', 'C-', 'C', 'C+', 'B-', 'B', 'B+', 'A-', 'A', 'A+'],
                                  ordered = True)

## Q7

```python
student_df = pd.DataFrame([{'First Name' : 'James', 'Last Name' : 'Hammond',
                          'School' : 'Business'},
                         {'First Name' : 'Mike', 'Last Name' : 'Smith',
                          'School' : 'Law'},
                         {'First Name' : 'Sally', 'Last Name' : 'Brooks',
                          'School' : 'Engineering'}])

staff_df = pd.DataFrame([{'First Name' : 'Kelly', 'Last Name' : 'Desjardins',
                          'Role' : 'Director of HR'},
                         {'First Name' : 'Sally', 'Last Name' : 'Brooks',
                          'Role' : 'Course liasion'},
                         {'First Name' : 'James', 'Last Name' : 'Wilde',
                          'Role' : 'Grader'}])
```

```python
df = pd.merge(staff_df, student_df, how = 'right', on = ['First Name', 'Last Name'])
df
```

```python
df = pd.merge(student_df, staff_df, how = 'inner', on = ['First Name', 'Last Name'])
df
```

```python
df = pd.merge(student_df, staff_df, how = 'right', on = ['First Name', 'Last Name'])
df
```

```python
df = pd.merge(staff_df, student_df, how = 'outer', on = ['First Name', 'Last Name'])
df
```

### Answer:
df = pd.merge(staff_df, student_df, how = 'right', on = ['First Name', 'Last Name'])
