# Worldbank data

```python
import numpy as np
```

```python
worldbank = np.genfromtxt('/Users/jonathansuarezcaceres/Downloads/1_Data Science/Worldbank_data/54a621bf-a339-4f0c-b617-0bb3bbbc9009_Data.csv',
                          dtype = None, delimiter = ',', skip_header = 1, names = ('Time', 'Time code',
                            'Country', 'Country code', 'Income growth', 'Income total', 'CO2',
                            'GDP Total', 'GDP Growth', 'Population total', 'Population growth',
                            'Poverty <1.90', 'Poverty <3.20', 'Unemployment rate'))

worldbank
```

```python
GrowthLowUnemployment = worldbank[worldbank['Unemployment_rate'] < 10]['GDP_Growth'].mean()
GrowthHighUnemployment = worldbank[worldbank['Unemployment_rate'] > 15]['GDP_Growth'].mean()

print(GrowthLowUnemployment)
print(GrowthHighUnemployment)
```

```python
worldbank.dtype.names
```

```python
PovAbove5 = worldbank[worldbank['Poverty_190'] > 5]
PovAbove5
```

```python
PovAbove5_GDP = PovAbove5['GDP_Growth'].mean()
PovAbove5_Pop = PovAbove5['Population_growth'].mean()
PovAbove5_Un = PovAbove5['Unemployment_rate'].mean()

print(PovAbove5_GDP)
print(PovAbove5_Pop)
print(PovAbove5_Un)
```

```python
PovBelow5 = worldbank[worldbank['Poverty_190'] < 5]
PovBelow5
```

```python
PovBelow5_GDP = PovBelow5['GDP_Growth'].mean()
PovBelow5_Pop = PovBelow5['Population_growth'].mean()
PovBelow_Un = PovBelow5['Unemployment_rate'].mean()

print(PovBelow5_GDP)
print(PovBelow5_Pop)
print(PovBelow_Un)
```

```python
from tabulate import tabulate

array = [['GDP Growth', PovAbove5_GDP, PovBelow5_GDP], ['Population Growth', PovAbove5_Pop, PovBelow5_Pop], 
        ['Unemployment', PovAbove5_Un, PovBelow_Un]]

headers = ['', 'Above 5', 'Below 5']
table = tabulate(array, headers, tablefmt = 'fancy_grid')
print(table)
```
